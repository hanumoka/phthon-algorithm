# 피아노 건반에서 최단거리 찾기

 ![피아노건반](img/피아노건반2.png)

# 위처럼 짧은 피아노 건반이 있다.
- 흰건반 : 1,3,5,7,8,10,12
- 검은건반: 2,4,6,9,11

조건
- 시작 말은 무조건 1 부터 시작한다.
- 말의 이동은 인접한 건반을 거쳐 이동할 수 있다.
- 말의 위치가 1인경우 2,3 으로 이동 가능
- 말의 위치가 3인 경우 1,2,4,5 이동 가능
- 말의 위치가 8인 경우 7, 9, 10 이동 가능

문제
[5, 7, 12, 2] 와 같이 건반 숫자 배열이 있는 경우 1부터 시작하는 말의 최단 이동거리를 리턴하라.

입력예시1 => [5, 7, 12, 2]

출력예시1 => 12

입력예시2 => [3, 4, 11, 5]

출력예시2 => 11

### 검사대상

```python
# 정방향
test_1 = { 'name': 'test_1', 'q': [5], 'a': 2}
test_2 = { 'name': 'test_2', 'q': [5, 8], 'a': 4}
test_6 = { 'name': 'test_6', 'q': [3,4,6,9], 'a': 7}
test_7 = { 'name': 'test_7', 'q': [2,4,6], 'a': 5}
test_8 = { 'name': 'test_8', 'q': [7,9,11,12], 'a': 8}

# 역방향
test_3 = { 'name': 'test_3', 'q': [5, 8, 1], 'a': 8}
test_4 = { 'name': 'test_4', 'q': [5,7,12,2], 'a': 12}
test_5 = { 'name': 'test_5', 'q': [3,4,11,5], 'a': 11}

test_list_a = [test_1, test_2, test_6, test_7, test_8]
test_list_b = [test_3, test_4, test_5]
test_list = test_list_a + test_list_b
```

### 솔루션

```python
def solution(q_array):
#     print(q_array)
    answer = 0
    
    piano_table = [1,0,1,0,1,0,1,1,0,1,0,1]
    
    before_val = 1;
    before_pos = 0;
    
    for target in q_array:
        
        if target > before_val:
            tmp_dis = 0
            for pos in range(before_pos + 1, len(piano_table)):
                if target == pos + 1:
                    tmp_dis = tmp_dis + 1
                    print("{} 에서 {}로 이동거리:{} 누적이동거리:{}".format(before_val, target, tmp_dis,answer + tmp_dis))
                    before_val = target
                    before_pos = pos
                    answer = answer + tmp_dis
                    break
                elif piano_table[pos] == 1:
                    tmp_dis = tmp_dis + 1
        else:
            tmp_dis = 0
            print("역박향 before_pos:{}".format(before_pos))
            for pos in range(before_pos - 1, -1, -1): # 거꾸로 배열 탐색
                if target == pos + 1:
                    tmp_dis = tmp_dis + 1
                    print("{} 에서 {}로 이동거리:{} 누적이동거리:{}".format(before_val, target, tmp_dis,answer + tmp_dis))
                    before_val = target
                    before_pos = pos
                    break
                elif piano_table[pos] == 1:
                    tmp_dis = tmp_dis + 1
            answer = answer + tmp_dis
            
    print('answer: {}'.format(answer))
    return answer
```

### 검사코드

```python
for test in test_list:
    print("==========================================")
    print(test)
    res = solution(test['q'])
    if res != test['a']:
        print(test['name'] + " 실패")
    else:
        print(test['name'] + " 성공")
```
